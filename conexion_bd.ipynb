```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine, text
import pandas as pd

# 1) Cargar variables de entorno
load_dotenv()

usuario = os.getenv("DB_USER")
password = os.getenv("DB_PASSWORD")
host = os.getenv("DB_HOST")
puerto = os.getenv("DB_PORT")
base = os.getenv("DB_NAME")

print("Base:", base, "- Usuario:", usuario)  # chequeo rápido

# 2) Crear engine de conexión
url = f"postgresql+psycopg2://{usuario}:{password}@{host}:{puerto}/{base}"
engine = create_engine(url)

# 3) Probar conexión
with engine.connect() as conn:
    resultado = conn.execute(text("SELECT 'Conexión OK a la base olist';"))
    print(resultado.scalar())
```
